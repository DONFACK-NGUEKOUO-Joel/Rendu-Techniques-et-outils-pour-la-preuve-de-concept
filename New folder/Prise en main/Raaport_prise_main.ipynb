### **1- Importation des bibliothèques et Lecture des données** 
<p style="text-align:justify"> Il est important d’importer les bibliothèques lors de l’exécution d’un code, car elles contiennent des fonctions et des outils déjà développés qui facilitent la programmation. Elles permettent de gagner du temps, d’éviter de réécrire des algorithmes complexes et d’assurer une meilleure fiabilité du code. Pour notre cas, les bibliothèques que nous avons utilisees sont les suivantes.
</p>

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
fichier = "data.txt"  
data = np.loadtxt(fichier) 
```

### **2- Vérification de la taille de la matrice du fichier** 
<p style="text-align:justify"> 
Il est important de vérifier la taille du fichier .txt, notamment la dimension de la matrice qu’il contient, afin de s’assurer que les données importées correspondent bien à ce que le programme attend. En effet, si la matrice est trop grande, le traitement peut devenir très lent ou même impossible à exécuter faute de mémoire. À l’inverse, une matrice trop petite ou incomplète risque de fausser les calculs.
</p>

```python
if data.shape[1] != 3:
    raise ValueError("Le fichier doit contenir 3 colonnes : x y z")

x = data[:, 0]
y = data[:, 1]
z = data[:, 2]
```

### **3- Calcul du plan des moindres carrés** 
Le modèle que nous avons adopté pour le calcul des moidre carré est le suivant: 

```python
A = np.c_[x, y, np.ones(x.shape)]
coeffs, _, _, _ = np.linalg.lstsq(A, z, rcond=None)
a, b, c = coeffs

print("✅ Équation du plan des moindres carrés :")
print(f"z = {a:.4f} * x + {b:.4f} * y + {c:.4f}") 
```

### **4- Visualisation du nuage des points**
Apres avoir calculé le nuage des point nous avons utilisé certaines fonctions prédéfinies afin d'afficher le résultat obtenu.

```python
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 

ax.scatter(x, y, z, color='black', label="Données")

ax.set_xlabel("X") 
ax.set_ylabel("Y")  
ax.set_zlabel("Z")
```

### **5- Ajustement du plan de visualisation**

<p style="text-align:justify"> 
Il est important de réaliser cette étape du plan ajusté car elle permet de visualiser graphiquement le plan obtenu par la méthode des moindres carrés. En créant des grilles de points avec
x_grid et y_grid, on génère un maillage régulier des valeurs de 
x et 𝑦 sur lequel on calcule les valeurs correspondantes de z à partir de l’équation du plan z= ax+by+c.  permet 
ensuite de tracer une surface continue représentant le plan ajusté par rapport aux points expérimentaux. Cette visualisation aide à vérifier la qualité de l’ajustement et à mieux interpréter les résultats obtenus.
</p>

### **6- Titres et Labels**
<p style="text-align:justify"> 
Les titres et labels (étiquettes des axes) sont très importants dans une représentation graphique, car ils permettent de comprendre facilement le contenu et la signification du graphique. Le titre donne une vue d’ensemble sur ce que le graphe représente (par exemple : “Plan des moindres carrés”), tandis que les labels sur les axes indiquent les variables et leurs unités, facilitant ainsi l’interprétation des données. Sans ces éléments, le graphique serait difficile à lire et pourrait prêter à confusion. Ils assurent donc une présentation claire, compréhensible et professionnelle des résultats.
</p>

```python
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 

ax.scatter(x, y, z, color='black', label="Données")
x_grid, y_grid = np.meshgrid(
    np.linspace(min(x), max(x), 20),
    np.linspace(min(y), max(y), 20)
)
z_grid = a * x_grid + b * y_grid + c
ax.plot_surface(x_grid, y_grid, z_grid, alpha=0.75, color='blue')

ax.set_xlabel("X") 
ax.set_ylabel("Y")  
ax.set_zlabel("Z")  
ax.set_title("Plan des moindres carrés") 
ax.legend()

plt.show()
```
